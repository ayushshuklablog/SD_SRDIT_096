### Add New Student Data

This section provides a function to manually add new student records, calculate their average marks, and append them to the existing DataFrame. You can customize the `subject_columns` list in the function definition if your subject names change.

```python


print("========== STUDENT MANAGEMENT SYSTEM ==========")


name = input("Enter Student Name: ")
roll = input("Enter Roll Number: ")

subjects = ["English", "Mathematics", "Science", "Social Studies", "Computer"]

marks = []


for subject in subjects:
    while True:
        mark = float(input(f"Enter marks for {subject}: "))
        if 0 <= mark <= 100:
            marks.append(mark)
            break
        else:
            print("Please enter marks between 0 and 100.")


total = sum(marks)
average = total / len(subjects)
if average >= 90:
    grade = "A+"
elif average >= 80:
    grade = "A"
elif average >= 70:
    grade = "B"
elif average >= 60:
    grade = "C"
elif average >= 50:
    grade = "D"
else:
    grade = "F"


print("\n")
print("=" * 45)
print("           STUDENT REPORT CARD")
print("=" * 45)
print(f"Student Name : {name}")
print(f"Roll Number  : {roll}")
print("-" * 45)

for i in range(len(subjects)):
    print(f"{subjects[i]:20} : {marks[i]}")

print("-" * 45)
print(f"Total Marks        : {total:.2f} / 500")
print(f"Average Marks      : {average:.2f}")
print(f"Grade              : {grade}")
print("=" * 45)
```
