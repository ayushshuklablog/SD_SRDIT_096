```python
def count_test(n):
  count = 1
  while count <= n :
    yield count
    count = count +1
```

```python
c = count_test(5)
```

```python
print(c)
```

```python
type(c)
```

```python
for i in c:
  print(i)
```

```python
def count_test1(n):
  count = 1
  while count <= n :
    yield count
    count = count *3
```

```python
c1 = count_test1(15)
```
