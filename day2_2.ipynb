```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b != 0:
        return a / b
    return "Cannot divide by zero"

def modulus(a, b):
    return a % b

def power(a, b):
    return a ** b

def floor_division(a, b):
    if b != 0:
        return a // b
    return "Cannot divide by zero"

def average(a, b):
    return (a + b) / 2


# Input
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

# Menu
print("\nChoose operation:")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")
print("5. Modulus (Remainder)")
print("6. Power (a^b)")
print("7. Floor Division")
print("8. Average")

choice = input("Enter your choice (1-8): ")

# Conditions
if choice == "1":
    print("Result =", add(a, b))
elif choice == "2":
    print("Result =", subtract(a, b))
elif choice == "3":
    print("Result =", multiply(a, b))
elif choice == "4":
    print("Result =", divide(a, b))
elif choice == "5":
    print("Result =", modulus(a, b))
elif choice == "6":
    print("Result =", power(a, b))
elif choice == "7":
    print("Result =", floor_division(a, b))
elif choice == "8":
    print("Result =", average(a, b))
else:
    print("Invalid choice")
```
